## Seasonality

- **Definition**: Repeating patterns or cycles in the data that occur at regular intervals.
- **Example**: Daily temperature fluctuations, yearly holiday shopping spikes.

Seasonality in time series data refers to the recurring patterns or trends that happen at specific intervals. Yearly holiday shopping spikes are a classic example of seasonality. In this example, I'll teach you how to analyze and visualize yearly holiday shopping spikes using Python and synthetic data. We'll use the `pandas`, `numpy`, and `matplotlib` libraries to create and visualize the data.

### Step 1: Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Step 2: Generate Synthetic Data

Generate synthetic data to simulate the yearly holiday shopping spikes. In this example, we'll create data for five years with monthly intervals:

```python
# Create a date range for five years with monthly intervals
date_rng = pd.date_range(start="2020-01-01", end="2024-12-31", freq="M")

# Create synthetic data with yearly holiday spikes
np.random.seed(0)
shopping_data = 100 + np.sin(np.arange(len(date_rng)) / 12 * 2 * np.pi) * 20 + np.random.normal(0, 5, len(date_rng))

# Create a DataFrame with the synthetic data
shopping_df = pd.DataFrame(shopping_data, columns=["Sales"])
shopping_df.set_index(date_rng, inplace=True)
```

```python
shopping_df.head()
```

### Step 3: Visualize the Data

Visualize the synthetic data to observe the yearly holiday shopping spikes:

```python
plt.figure(figsize=(12, 6))
plt.plot(shopping_df, label="Monthly Sales")
plt.title("Yearly Holiday Shopping Spikes")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

### Step 4: Analyze Seasonality

To analyze the seasonality in the data more formally, you can calculate the mean sales for each month over the five-year period:

```python
# Calculate the mean sales for each month
monthly_mean = shopping_df.groupby(shopping_df.index.month).mean()

# Visualize the monthly means
plt.figure(figsize=(12, 6))
plt.plot(monthly_mean, marker='o')
plt.title("Monthly Mean Sales Over Five Years")
plt.xlabel("Month")
plt.ylabel("Mean Sales")
plt.xticks(np.arange(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()

```

You should observe the highest mean sales during the holiday months (e.g., November and December), reflecting the yearly holiday shopping spikes.

This example demonstrates how to create synthetic data and analyze the seasonality of yearly holiday shopping spikes. In real-world scenarios, you'd apply similar techniques to analyze actual sales data and make data-driven decisions based on seasonality.
